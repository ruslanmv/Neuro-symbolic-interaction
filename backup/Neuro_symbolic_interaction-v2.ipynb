```python
import os
import shutil
from IPython.display import clear_output
from google.colab import userdata
# Retrieve the GitHub token from Colab secrets
GITHUB_TOKEN = userdata.get('GITHUB_TOKEN_PERSONAL')
# Repository URL and name
repo_url = "https://github.com/ruslanmv/Neuro-symbolic-interaction.git"
repo_name = "Neuro-symbolic-interaction"

# Create the clone command with the token
clone_command = f"git clone https://{GITHUB_TOKEN}@{repo_url.replace('https://', '')}"

# Check if the repository directory exists
if not os.path.exists(repo_name):
    print(f"Cloning {repo_name} repository...")
    !{clone_command}
    clear_output()
    print(f"{repo_name} repository cloned successfully!")

    # Move contents of the repository to the current directory
    print(f"Moving contents of {repo_name} to current directory...")
    for item in os.listdir(repo_name):
        s = os.path.join(repo_name, item)
        d = os.path.join(".", item)  # Current directory
        try:
            shutil.move(s, d)  # Move files and directories
        except Exception as e:
            print(f"Error moving {item}: {e}")

    shutil.rmtree(repo_name)  # Remove the now-empty repo directory
    print(f"Contents of {repo_name} moved successfully!")
else:
    print(f"{repo_name} repository already exists. Skipping cloning.")
print("Finished.")
```

```python
import json
import pandas as pd
from backend import handle_evaluation  # Import the evaluation function from backend.py

# Load questions and true answers from questions.json
def load_questions_from_json(json_file_path):
    with open(json_file_path, 'r') as f:
        question_answer_pairs = json.load(f)
    questions = list(question_answer_pairs.keys())
    true_answers = list(question_answer_pairs.values())
    return questions, true_answers

def perform_inference(question, use_ontology, ontology_path=None):
    """
    Performs inference for a given question using the handle_evaluation function.

    Args:
        question (str): The question to ask.
        use_ontology (bool): Whether to use ontology during inference.
        ontology_path (str, optional): Path to the ontology file. Defaults to None (using default ontology).

    Returns:
        str: The LLM's answer (extracted statement from the JSON response).
    """
    response_json_str = handle_evaluation(question, use_ontology, ontology_path)
    response_data = json.loads(response_json_str)
    llm_answer = response_data.get("statements", "No answer found.") # Extract 'statements' as answer
    return llm_answer

def main():
    json_file_path = "questions.json"
    questions, true_answers = load_questions_from_json(json_file_path)
    ontology_path = "engine_ontology.owl" # Path to your ontology file

    results = []

    for i in range(len(questions)):
        question = questions[i]
        true_answer = true_answers[i]

        # Inference without ontology
        llm_answer_no_ontology = perform_inference(question, use_ontology=False)

        # Inference with ontology
        llm_answer_with_ontology = perform_inference(question, use_ontology=True, ontology_path=ontology_path)

        results.append({
            "Question": question,
            "LLM Answer": llm_answer_no_ontology,
            "LLM Answer with Ontology": llm_answer_with_ontology,
            "True Answer": true_answer
        })

    # Create Pandas DataFrame
    df = pd.DataFrame(results)

    # Save DataFrame to CSV file
    csv_file_path = "inference_results.csv"
    df.to_csv(csv_file_path, index=False)

    print(f"Inference results saved to '{csv_file_path}'")

if __name__ == "__main__":
    main()
```
