```python
import csv
import json
import os
import pandas as pd
import matplotlib.pyplot as plt
from evaluation.metrics import (
    calculate_bleu, calculate_rouge, calculate_coherence, calculate_hrr,
    calculate_lcs, calculate_ra, evaluate_knowledge_retention
)
from backend import handle_evaluation, main
#from evaluation.data import evaluation_data


def log_evaluation_results(logs_path, question, statement, knowledge_base, label, ontology):
    """
    Logs evaluation results into a pickle file.

    Parameters:
    - logs_path: Path to the pickle file for logging.
    - question: The original user question.
    - statement: The generated statement or response.
    - knowledge_base: The knowledge base or logical form evaluated.
    - label: Whether the statement is true or false.
    - ontology: Whether ontology was used ('True' or 'False').
    """
    # Create a dictionary with the data
    data = {
        'Question': question,
        'Statement': statement,
        'Knowledge Base': knowledge_base,
        'Label': label,
        'Ontology': ontology
    }

    # Load existing DataFrame if the file exists, otherwise create a new one
    if os.path.exists(logs_path):
        df = pd.read_pickle(logs_path)
    else:
        df = pd.DataFrame(columns=data.keys())

    # Append the new data to the DataFrame
    df = pd.concat([df, pd.DataFrame([data])], ignore_index=True)

    # Save the updated DataFrame to the pickle file
    df.to_pickle(logs_path)


def evaluate_model(ontology_path, statements, gold_answers, use_ontology, logs_path="logs.pkl"):
    """
    Evaluates the model performance and logs the results.

    Parameters:
    - ontology_path: Path to the ontology file.
    - statements: List of input queries/statements.
    - gold_answers: List of gold answers corresponding to each query.
    - use_ontology: Boolean flag to toggle ontology usage.
    - logs_path: Path to the logs file (default: "logs.pkl").
    """
    # Initialize ontology, model, and vectorizer using the main function
    onto, model, vectorizer = main(ontology_path)

    results = []
    baseline_outputs = [{"query": s, "is_hallucination": True} for s in statements]
    validated_outputs = [{"query": s, "is_hallucination": False} for s in statements]

    for i, statement in enumerate(statements):
        print(f"Evaluating statement {i+1}: {statement} (Ontology: {'Enabled' if use_ontology else 'Disabled'})")
        response = handle_evaluation(statement, use_ontology, ontology_path)

        # Parse response from JSON string into a dictionary
        response_dict = json.loads(response)

        query = statement
        output = response_dict.get('statements', '') or "No response generated."
        # Use the corresponding gold answer from the provided list
        gold_answer = gold_answers[i]

        # Determine label based on whether a response was generated
        label = 'True' if response_dict.get('statements') else 'False'

        # Log the evaluation details
        log_evaluation_results(
            logs_path=logs_path,
            question=query,
            statement=output,
            knowledge_base=response_dict.get('details', 'N/A'),
            label=label,
            ontology=str(use_ontology)
        )

        # Metrics evaluation
        bleu_score = calculate_bleu(output, gold_answer)
        rouge_scores = calculate_rouge(output, gold_answer)
        coherence_score = calculate_coherence(query, output)
        hrr_score = calculate_hrr(baseline_outputs, validated_outputs)
        #lcs_score = calculate_lcs(validated_outputs)
        lcs_score = calculate_lcs(output, gold_answer)
        ra_score = calculate_ra([{"correct_answer": gold_answer}], [{"output": output}])
        knowledge_retention_score = evaluate_knowledge_retention([query], [gold_answer], [output])

        # Append results for the current evaluation
        results.append({
            "Example ID": f"{use_ontology}-{i + 1}",
            "Query": query,
            "Gold Answer": gold_answer,
            "Model Output": output,
            "Use Ontology": use_ontology,
            "BLEU Score": round(bleu_score, 2),
            "ROUGE-1 F1": round(rouge_scores['rouge1'].fmeasure, 2),
            "ROUGE-L F1": round(rouge_scores['rougeL'].fmeasure, 2),
            "Coherence Score": round(coherence_score, 2),
            "HRR": hrr_score,
            "LCS": lcs_score,
            "RA": ra_score,
            "Knowledge Retention": round(knowledge_retention_score, 2),
        })

    return results


def save_results(results, output_csv="evaluation_results.csv", latex_file="results.tex"):
    """
    Saves the evaluation results to a CSV file and generates LaTeX reports.

    Parameters:
    - results: List of dictionaries containing evaluation metrics.
    - output_csv: Filename for saving the CSV output.
    - latex_file: Filename for saving the LaTeX report.
    """
    df = pd.DataFrame(results)
    # Debug: Print DataFrame before saving
    print("Results DataFrame:\n", df.head(), "\n")
    df.to_csv(output_csv, index=False)
    print(f"Results saved to {output_csv}")


# Example usage:
# Assume you have a pandas DataFrame `df` with columns 'question' and 'gold_answer'.
# For example:
question_list = df['question'].tolist()
gold_answer_list = df['gold_answer'].tolist()

ontology_path = "engine_ontology.owl"

# Example: Evaluate only the first question/answer pair for testing purposes.
results_without_ontology = evaluate_model(
    ontology_path,
    question_list[:1],
    gold_answer_list[:1],
    use_ontology=False
)
```

```python
"""
advanced_metrics.py

Provides advanced evaluation metrics for neurosymbolic or ontology-based systems:
1) Factual Accuracy
2) Consistency Score
3) Efficiency and Latency

Note: Metrics like Ontology Alignment, Error Detection/Correction, and Human Evaluation
are not automatically computed in this version as they require external data sources or human input.

Author: Your Name
Date: 2023-12-08
"""

import time
import pandas as pd
from typing import Callable

# ------------------------------------------------------------------------
# 5.1 Factual Accuracy
# ------------------------------------------------------------------------
def calculate_factual_accuracy(model_output: str, gold_answer: str) -> float:
    """
    Calculates factual accuracy as the percentage token overlap between
    model_output and gold_answer.
    """
    # Ensure inputs are strings
    if not isinstance(model_output, str):
        model_output = str(model_output)
    if not isinstance(gold_answer, str):
        gold_answer = str(gold_answer)

    model_tokens = set(model_output.lower().split())
    gold_tokens = set(gold_answer.lower().split())

    if not gold_tokens:
        return 0.0
    overlap = model_tokens.intersection(gold_tokens)
    return (len(overlap) / len(gold_tokens)) * 100

# ------------------------------------------------------------------------
# 5.2 Consistency Score
# ------------------------------------------------------------------------
def calculate_consistency_score(model_output: str) -> float:
    """
    Evaluates the internal logical consistency of the model's output.
    This heuristic reduces the score by 5% for each occurrence of contradictory keywords.
    """
    # Ensure input is a string
    if not isinstance(model_output, str):
        model_output = str(model_output)

    contradictory_keywords = ["contradiction", "but", "however", "although", "nevertheless"]
    count = sum(model_output.lower().count(kw) for kw in contradictory_keywords)
    # Each contradictory keyword reduces consistency by 5%
    score = max(0.0, 1.0 - count * 0.05)
    return score

# ------------------------------------------------------------------------
# 5.6 Efficiency and Latency
# ------------------------------------------------------------------------
def measure_efficiency_and_latency(function_to_run: Callable) -> float:
    """
    Measures the execution time of a given function call.
    """
    start_time = time.time()
    function_to_run()
    end_time = time.time()
    return end_time - start_time

def dummy_processing(model_output: str):
    """
    Simulates processing time based on the length of model_output.
    For example, it sleeps for 0.0005 seconds per word.
    """
    # Ensure input is a string
    if not isinstance(model_output, str):
        model_output = str(model_output)

    sleep_time = 0.0005 * len(model_output.split())
    time.sleep(sleep_time)

# ------------------------------------------------------------------------
# Main computation function
# ------------------------------------------------------------------------
def compute_advanced_metrics(csv_file="evaluation_results.csv", updated_file="evaluation_results.csv"):
    """
    Reads the CSV file, computes advanced metrics, and saves the updated results.
    """
    # 1) Load the CSV (accepting either a file path or a DataFrame)
    try:
        if isinstance(csv_file, pd.DataFrame):
            df = csv_file.copy()
        else:
            df = pd.read_csv(csv_file)
    except Exception as e:
        print(f"Error reading CSV file: {e}")
        return

    # Ensure required columns exist
    if "Gold Answer" not in df.columns or "Model Output" not in df.columns:
        print("Error: CSV file must contain 'Gold Answer' and 'Model Output' columns.")
        return

    # 2) Initialize lists to store computed metrics
    factual_acc_list = []
    consistency_score_list = []
    latency_list = []

    # 3) Iterate over each row and compute metrics
    for idx, row in df.iterrows():
        gold_answer = row.get("Gold Answer", "")
        model_output = row.get("Model Output", "")

        # Compute factual accuracy (0-100)
        factual_acc = calculate_factual_accuracy(model_output, gold_answer)

        # Compute consistency score (0.0 to 1.0)
        consistency_score = calculate_consistency_score(model_output)

        # Measure latency with dummy processing that varies with the length of model_output
        latency = measure_efficiency_and_latency(lambda: dummy_processing(model_output))

        # Append computed metrics (rounded for readability)
        factual_acc_list.append(round(factual_acc, 2))
        consistency_score_list.append(round(consistency_score, 2))
        latency_list.append(round(latency, 4))

    # 4) Insert new metric columns into the DataFrame
    df["Factual Accuracy"] = factual_acc_list
    df["Consistency Score"] = consistency_score_list
    df["Latency (sec)"] = latency_list

    # 5) Save updated results to CSV
    try:
        df.to_csv(updated_file, index=False)
        print(f"Updated evaluation results saved to {updated_file}")
    except Exception as e:
        print(f"Error saving updated CSV file: {e}")


```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import pickle


# Process evaluation results
def process_results(input_csv="evaluation_results.csv", results_dir="results"):
    # Ensure results directory exists
    os.makedirs(results_dir, exist_ok=True)

    # Read evaluation results
    df = pd.read_csv(input_csv)
    print(f"Read {len(df)} rows from {input_csv}\n")

    # Debug: Print the raw data
    print("Original DataFrame:\n", df.head().values, "\n")

    # Separate results with and without ontology
    df_with_ontology = df[df["Use Ontology"] == True].copy()
    df_without_ontology = df[df["Use Ontology"] == False].copy()

    # Debug: Print individual filtered DataFrames
    print("df_with_ontology:\n", df_with_ontology.head(1).values, "\n")
    print("df_without_ontology:\n", df_without_ontology.head(1).values, "\n")

    # Add a consistent merge key based on the Example ID's number (strip prefixes like "True-" or "False-")
    df_with_ontology["Example ID"] = df_with_ontology["Example ID"].str.replace(r"^(True-|False-)", "", regex=True)
    df_without_ontology["Example ID"] = df_without_ontology["Example ID"].str.replace(r"^(True-|False-)", "", regex=True)

    # Use Example ID as the merge key
    df_with_ontology["Merge Key"] = df_with_ontology["Example ID"]
    df_without_ontology["Merge Key"] = df_without_ontology["Example ID"]

    # Debug: Check unique keys
    print("Unique Merge Keys in df_with_ontology:\n", df_with_ontology["Merge Key"].unique(), "\n")
    print("Unique Merge Keys in df_without_ontology:\n", df_without_ontology["Merge Key"].unique(), "\n")

    # Merge results for comparison
    common_cols = [
        "ROUGE-1 F1", "ROUGE-L F1", "Coherence Score",
        "HRR", "LCS", "RA", "Knowledge Retention"
    ]
    # Convert metrics to numeric, coerce errors to NaN
    for col in common_cols:
        df_with_ontology[col] = pd.to_numeric(df_with_ontology[col], errors="coerce")
        df_without_ontology[col] = pd.to_numeric(df_without_ontology[col], errors="coerce")

    merged_df = pd.merge(
        df_with_ontology[["Merge Key", "Example ID"] + common_cols],
        df_without_ontology[["Merge Key", "Example ID"] + common_cols],
        on="Merge Key",
        suffixes=("_With_Ontology", "_Without_Ontology"),
        how="inner"
    )

    # Debug: Print the merged DataFrame or a message if it's empty
    if merged_df.empty:
        print("Merged DataFrame is empty. Ensure that `Example ID` values match in the input data.")
        return
    else:
        print("Merged DataFrame:\n", merged_df.head(), "\n")

    # Calculate differences between metrics
    for metric in common_cols:
        merged_df[f"{metric} Difference"] = (
            merged_df[f"{metric}_With_Ontology"] - merged_df[f"{metric}_Without_Ontology"]
        )

    # Save merged results
    merged_csv = os.path.join(results_dir, "merged_results.csv")
    merged_df.to_csv(merged_csv, index=False)
    print(f"Merged results saved to {merged_csv}\n")

    # Save as a pickle file
    merged_pickle = os.path.join(results_dir, "merged_results.pkl")
    with open(merged_pickle, "wb") as f:
        pickle.dump(merged_df, f)
    print(f"Merged results saved to {merged_pickle}\n")

    # Generate plots for comparison
    generate_comparison_plots(merged_df, results_dir)

    # Generate LaTeX report
    latex_file = os.path.join(results_dir, "results.tex")
    generate_latex_report(merged_df, filename=latex_file, results_dir=results_dir)
    print(f"LaTeX report saved to {latex_file}")

    # Create a legend mapping Merge Key to Query
    legend = pd.concat([df_with_ontology[["Merge Key", "Query"]], df_without_ontology[["Merge Key", "Query"]]]).drop_duplicates()
    legend_csv = os.path.join(results_dir, "legend.csv")
    legend.to_csv(legend_csv, index=False)
    print(f"Legend saved to {legend_csv}\n")


# Generate comparison plots
def generate_comparison_plots(df, results_dir):
    metrics = [
        "ROUGE-1 F1", "ROUGE-L F1", "Coherence Score",
        "HRR", "LCS", "RA", "Knowledge Retention"
    ]

    for metric in metrics:
        plt.figure(figsize=(12, 6))

        # Drop rows with NaN values for the current metric
        plot_df = df.dropna(subset=[f"{metric}_With_Ontology", f"{metric}_Without_Ontology"])

        # Use numeric keys for x-axis labels
        x_labels = plot_df["Merge Key"].astype(int).tolist()

        if plot_df.empty:
            print(f"No data available for plotting {metric}. Skipping plot.\n")
            continue

        plt.plot(x_labels, plot_df[f"{metric}_With_Ontology"], label=f"{metric} With Ontology", marker='o')
        plt.plot(x_labels, plot_df[f"{metric}_Without_Ontology"], label=f"{metric} Without Ontology", marker='x')
        plt.fill_between(
            x_labels,
            plot_df[f"{metric}_With_Ontology"],
            plot_df[f"{metric}_Without_Ontology"],
            color='gray', alpha=0.3, label=f"Difference"
        )
        plt.title(f"{metric} Comparison: With vs Without Ontology", fontsize=14)
        plt.xlabel("Question Number", fontsize=12)
        plt.ylabel(metric, fontsize=12)
        plt.xticks(x_labels, x_labels, rotation=0)
        plt.legend()
        plt.grid(True)
        plt.tight_layout()

        plot_file = os.path.join(results_dir, f"{metric.replace(' ', '_')}_Comparison.pdf")
        plt.savefig(plot_file)
        print(f"Comparison plot saved to {plot_file}\n")


# Generate a comprehensive LaTeX report
import os

# Generate a comprehensive LaTeX report
def generate_latex_report(df, filename="results.tex", results_dir="results"):
    latex_table = """
\\documentclass{article}
\\usepackage{graphicx}
\\usepackage{longtable}
\\usepackage{geometry}
\\geometry{a4paper, margin=1in}
\\title{Experimental Results}
\\author{}
\\date{}

\\begin{document}
\\maketitle
\\section*{Evaluation Metrics Across Examples}
\\begin{longtable}{|c|l|l|l|l|l|l|l|l|}
\\hline
\\textbf{Example ID} & \\textbf{BLEU} & \\textbf{ROUGE-1 F1} & \\textbf{ROUGE-L F1} & \\textbf{Coherence} & \\textbf{HRR} & \\textbf{LCS} & \\textbf{RA} & \\textbf{Knowledge Retention} \\\\ \\hline
"""
    for _, row in df.iterrows():
        latex_table += f"{row['Example ID_With_Ontology']} & {row['ROUGE-1 F1_With_Ontology']:.2f} & {row['ROUGE-1 F1_Without_Ontology']:.2f} & {row['ROUGE-L F1_With_Ontology']:.2f} & {row['Coherence Score_With_Ontology']:.2f} & {row['HRR_With_Ontology']:.2f} & {row['LCS_With_Ontology']:.2f} & {row['RA_With_Ontology']:.2f} & {row['Knowledge Retention_With_Ontology']:.2f} \\\\ \\hline\n"

    latex_table += """
\\end{longtable}

\\section*{Plots of Evaluation Metrics}

"""
    metrics = [
        "ROUGE-1 F1", "ROUGE-L F1", "Coherence Score",
        "HRR", "LCS", "RA", "Knowledge Retention"
    ]
    for metric in metrics:
        plot_file = os.path.join(results_dir, f"{metric.replace(' ', '_')}_Comparison.pdf")
        latex_table += f"""
\\begin{{figure}}[h!]
\\centering
\\includegraphics[width=0.8\\textwidth]{{{{{plot_file}}}}}
\\caption{{{metric} Across Examples}}
\\end{{figure}}
"""

    latex_table += """
\\end{document}
"""

    with open(filename, "w") as file:
        file.write(latex_table)
    print(f"LaTeX file saved as {filename}")


# Run the processing function
process_results(input_csv="evaluation_results.csv")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# --- Read Data from CSV ---
# Make sure your CSV file is named 'evaluation_results.csv'
data = pd.read_csv("evaluation_results.csv")

# Map the boolean 'Use Ontology' to a string column 'Method' for clarity
data["Method"] = data["Use Ontology"].map({True: "With Ontology", False: "Without Ontology"})

# Optional: Print unique methods to verify the mapping
print("Unique Methods:", data["Method"].unique())

# -----------------------------------------------------
# Plot 1: Boxplot of BLEU Score by Method
# -----------------------------------------------------
fig, axs = plt.subplots(2, 2, figsize=(14, 12))

sns.boxplot(x="Method", y="BLEU Score", data=data, ax=axs[0, 0], palette="Set2")
axs[0, 0].set_title("Boxplot of BLEU Score by Method")
axs[0, 0].set_xlabel("")
axs[0, 0].set_ylabel("BLEU Score")

# -----------------------------------------------------
# Plot 2: Histogram with KDE for BLEU Score by Method
# -----------------------------------------------------
sns.histplot(data=data, x="BLEU Score", hue="Method", kde=True, ax=axs[0, 1],
             palette="Set2", alpha=0.6, bins=20)
axs[0, 1].set_title("Histogram of BLEU Score by Method")
axs[0, 1].set_xlabel("BLEU Score")
axs[0, 1].set_ylabel("Frequency")

# -----------------------------------------------------
# Plot 3: Scatter Plot of HRR vs LCS colored by Method
# -----------------------------------------------------
sns.scatterplot(x="HRR", y="LCS", hue="Method", data=data, ax=axs[1, 0],
                palette="Set2", s=80)
axs[1, 0].set_title("Scatter Plot: HRR vs LCS")
axs[1, 0].set_xlabel("Hallucination Reduction Rate (HRR)")
axs[1, 0].set_ylabel("Logical Consistency Score (LCS)")

# -----------------------------------------------------
# Plot 4: Bar Plot of Mean Evaluation Metrics by Method
# (Metrics: HRR, LCS, and Response Accuracy (RA))
# -----------------------------------------------------
metrics = ["HRR", "LCS", "RA"]
# Group by method and calculate mean and standard deviation
grouped = data.groupby("Method")[metrics].agg(["mean", "std"]).reset_index()

# Prepare a dataframe for plotting
plot_data = []
for method in grouped["Method"]:
    for metric in metrics:
        mean_val = grouped[grouped["Method"] == method][(metric, "mean")].values[0]
        std_val = grouped[grouped["Method"] == method][(metric, "std")].values[0]
        plot_data.append({"Method": method, "Metric": metric, "Mean": mean_val, "Std": std_val})
df_plot = pd.DataFrame(plot_data)

sns.barplot(x="Metric", y="Mean", hue="Method", data=df_plot, ax=axs[1, 1],
            palette="Set2", capsize=0.1)
axs[1, 1].set_title("Mean Evaluation Metrics by Method")
axs[1, 1].set_ylabel("Mean Score")
axs[1, 1].set_xlabel("Metric")

plt.tight_layout()
plt.show()

# -----------------------------------------------------
# Plot 5: Correlation Heatmap for Evaluation Metrics
# (Selected metrics: BLEU Score, ROUGE-1 F1, ROUGE-L F1, Coherence Score, HRR, LCS, RA, Knowledge Retention)
# -----------------------------------------------------
metrics_corr = ["BLEU Score", "ROUGE-1 F1", "ROUGE-L F1", "Coherence Score", "HRR", "LCS", "RA", "Knowledge Retention"]
corr_matrix = data[metrics_corr].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Evaluation Metrics")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

def plot_hallucination_vs_length():
    """
    Figure 1:
    Simulate how the hallucination rate (in %) increases with response length
    for a baseline LLM, and how the neurosymbolic approach keeps it lower.
    """
    # Simulated data for response lengths (in tokens)
    lengths = np.linspace(10, 100, 10)
    # Baseline LLM: hallucination rate increases from 10% to 70%
    baseline_hallucination = np.linspace(10, 70, 10)
    # Neurosymbolic approach: hallucination rate remains lower (from 5% to 20%)
    neurosymbolic_hallucination = np.linspace(5, 20, 10)

    plt.figure(figsize=(8, 6))
    plt.plot(lengths, baseline_hallucination, marker='o', linestyle='-', label='Baseline LLM', color='red')
    plt.plot(lengths, neurosymbolic_hallucination, marker='s', linestyle='--', label='Neurosymbolic Approach', color='green')
    plt.xlabel('Response Length (tokens)')
    plt.ylabel('Hallucination Rate (%)')
    plt.title("Figure 1: Hallucination Rate vs Response Length")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig("figure1_hallucination_vs_length.png")
    plt.show()

def plot_neurosymbolic_workflow():
    """
    Figure 2:
    Create a schematic diagram of the neurosymbolic workflow using a directed graph.
    """
    # Define the workflow steps
    steps = ["LLM Generation", "Formalization", "Ontology-Based Evaluation", "Iterative Refinement", "Final Output"]

    # Create a directed graph
    G = nx.DiGraph()
    for step in steps:
        G.add_node(step)
    # Define the workflow edges
    edges = [("LLM Generation", "Formalization"),
             ("Formalization", "Ontology-Based Evaluation"),
             ("Ontology-Based Evaluation", "Iterative Refinement"),
             ("Iterative Refinement", "Final Output")]
    G.add_edges_from(edges)

    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(G, seed=42)  # fixed layout for consistency
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=3000,
            arrowstyle='->', arrowsize=20, font_size=10, edge_color='gray')
    plt.title("Figure 2: Neurosymbolic Workflow Diagram")
    plt.tight_layout()
    plt.savefig("figure2_neurosymbolic_workflow.png")
    plt.show()

def plot_evaluation_metrics():
    """
    Figure 3:
    Generate a bar chart comparing the improvements in key evaluation metrics:
    - Hallucination Reduction Rate (HRR)
    - Logical Consistency Score (LCS) Improvement
    - Response Accuracy (RA) Improvement
    """
    # Define the metrics and corresponding improvement percentages
    metrics = ["Hallucination\nReduction Rate", "Logical Consistency\nImprovement", "Response Accuracy\nImprovement"]
    improvements = [78, 85, 63]  # in percentages

    plt.figure(figsize=(8, 6))
    bars = plt.bar(metrics, improvements, color=['#FF9999','#99FF99','#9999FF'])
    plt.ylim(0, 100)
    plt.ylabel("Improvement (%)")
    plt.title("Figure 3: Evaluation Metrics Improvements")

    # Annotate each bar with its percentage value
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2.0, height + 1, f'{height:.0f}%',
                 ha='center', va='bottom', fontsize=10)

    plt.tight_layout()
    plt.savefig("figure3_evaluation_metrics.png")
    plt.show()

def plot_ui_non_activated():
    """
    Figure 4:
    Simulate a dummy UI screenshot for the system when the neurosymbolic workflow is not activated.
    """
    plt.figure(figsize=(8, 6))
    plt.axis('off')
    plt.text(0.5, 0.5, "User Interface: Neurosymbolic Workflow NOT Activated",
             fontsize=16, ha='center', va='center', color='red',
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=1'))
    plt.title("Figure 4: UI with Neurosymbolic Workflow Not Selected")
    plt.tight_layout()
    plt.savefig("figure4_ui_non_activated.png")
    plt.show()

def plot_ui_activated():
    """
    Figure 5:
    Simulate a dummy UI screenshot for the system when the neurosymbolic workflow is activated.
    """
    plt.figure(figsize=(8, 6))
    plt.axis('off')
    plt.text(0.5, 0.5, "User Interface: Neurosymbolic Workflow ACTIVATED",
             fontsize=16, ha='center', va='center', color='green',
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=1'))
    plt.title("Figure 5: UI with Neurosymbolic Workflow Activated")
    plt.tight_layout()
    plt.savefig("figure5_ui_activated.png")
    plt.show()

def main():
    # Generate all plots sequentially
    plot_hallucination_vs_length()
    plot_neurosymbolic_workflow()
    plot_evaluation_metrics()
    plot_ui_non_activated()
    plot_ui_activated()


main()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Read Data from CSV ---
data = pd.read_csv("evaluation_results.csv")
print("Columns in the dataset:", data.columns.tolist())

# Create an 'Approach' column based on 'Use Ontology'
# Adjust the lambda function if your CSV uses other values (e.g., "Yes"/"No")
data['Approach'] = data['Use Ontology'].apply(
    lambda x: "Neurosymbolic" if x in [True, 1, "Yes", "yes"] else "LLM"
)

# Set seaborn style for better aesthetics
sns.set(style="whitegrid", font_scale=1.1)

#############################################
# Plot 1: Bar Plots for Average Metrics by Approach
#############################################
# Here, we use 'Coherence Score' (as Logical Consistency)
# and 'HRR' (as a proxy for Hallucination Rate)

grouped = data.groupby('Approach')[['Coherence Score', 'HRR']].mean().reset_index()

fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Bar plot for Coherence Score
sns.barplot(x='Approach', y='Coherence Score', data=grouped, ax=axs[0], palette="Blues_d")
axs[0].set_title("Average Coherence Score by Approach")
axs[0].set_ylabel("Coherence Score")
axs[0].set_xlabel("Approach")

# Bar plot for HRR
sns.barplot(x='Approach', y='HRR', data=grouped, ax=axs[1], palette="Reds_d")
axs[1].set_title("Average HRR by Approach")
axs[1].set_ylabel("HRR")
axs[1].set_xlabel("Approach")

plt.tight_layout()
plt.show()

#############################################
# Plot 2: Scatter Plot - Coherence Score vs. HRR
#############################################
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Coherence Score', y='HRR', hue='Approach',
                style='Approach', data=data, s=100)
plt.title("Coherence Score vs. HRR by Approach")
plt.xlabel("Coherence Score")
plt.ylabel("HRR")
plt.legend(title='Approach')
plt.show()

#############################################
# Plot 3: Line Plot of Metrics Across Test Cases
#############################################
# Use 'Example ID' as a proxy for test case ordering.
# This plot shows the trend of Coherence Score and HRR over examples for each approach.
plt.figure(figsize=(10, 6))
for approach in data['Approach'].unique():
    subset = data[data['Approach'] == approach].sort_values('Example ID')
    plt.plot(subset['Example ID'], subset['Coherence Score'], marker='o',
             label=f"{approach} Coherence Score")
    plt.plot(subset['Example ID'], subset['HRR'], marker='x',
             label=f"{approach} HRR")
plt.xlabel("Example ID")
plt.ylabel("Metric Value")
plt.title("Metrics Across Examples")
plt.legend()
plt.show()

#############################################
# Plot 4: Correlation Heatmap of Metrics
#############################################
# Select a set of numeric metric columns available in your CSV.
metrics_columns = ['BLEU Score', 'ROUGE-1 F1', 'ROUGE-L F1', 'Coherence Score',
                   'HRR', 'LCS', 'RA', 'Knowledge Retention', 'Response_Length']

# Ensure all selected columns exist in the DataFrame
available_metrics = [col for col in metrics_columns if col in data.columns]

plt.figure(figsize=(8, 6))
corr = data[available_metrics].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Metrics")
plt.show()

#############################################
# Plot 5: Box Plots for Metric Distributions by Approach
#############################################
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

sns.boxplot(x='Approach', y='Coherence Score', data=data, ax=axs[0], palette="Blues")
axs[0].set_title("Distribution of Coherence Score by Approach")
axs[0].set_xlabel("Approach")
axs[0].set_ylabel("Coherence Score")

sns.boxplot(x='Approach', y='HRR', data=data, ax=axs[1], palette="Reds")
axs[1].set_title("Distribution of HRR by Approach")
axs[1].set_xlabel("Approach")
axs[1].set_ylabel("HRR")

plt.tight_layout()
plt.show()
```
